```python
import numpy as np
import pandas as pd
```

```python
movies = pd.read_csv("Desktop/imdb_top_1000.csv")
```

```python
movies = movies[['Series_Title','Overview','Genre','Director','Star1','Star2','Star3','Star4']]
movies
```

```python
movies.dropna(inplace=True)
```

```python
movies.sample(7)
```

```python
movies['Director'] = movies['Director'].apply(lambda x: x.replace(" ", ""))
```

```python
movies.sample(5)
```

```python
movies['Star1'] = movies['Star1'].apply(lambda x: x.replace(" ", ""))
movies['Star2'] = movies['Star2'].apply(lambda x: x.replace(" ", ""))
```

```python
movies
```

```python
movies['Star3'] = movies['Star3'].apply(lambda x: x.replace(" ", ""))
movies['Star4'] = movies['Star4'].apply(lambda x: x.replace(" ", ""))
```

```python
movies
```

```python
movies['tags'] = (
    movies['Overview'] + " " +
    movies['Genre'] + " " +
    movies['Director'] + " " +
    movies['Star1'] + " " +
    movies['Star2'] + " " +
    movies['Star3'] + " " +
    movies['Star4']
)
```

```python
movies.sample(20)
```

```python
new = movies[['Series_Title','tags']]
```

```python
new.loc[:, 'tags'] = new['tags'].str.lower()
```

```python
from sklearn.feature_extraction.text import CountVectorizer

cv = CountVectorizer(max_features=5000, stop_words='english')
vectors = cv.fit_transform(new['tags']).toarray()
```

```python
from sklearn.metrics.pairwise import cosine_similarity

similarity = cosine_similarity(vectors)
```

```python
def recommend(movie):
    movie_index = new[new['Series_Title'] == movie].index[0]
    distances = similarity[movie_index]

    movies_list = sorted(
        list(enumerate(distances)),
        reverse=True,
        key=lambda x: x[1]
    )[1:6]

    for i in movies_list:
        print(new.iloc[i[0]].Series_Title)
```

```python
recommend("The Dark Knight")
```

```python
recommend("Iron Man")
```

```python
recommend("Sholay")
```

```python
recommend("After Hours")
```
